```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
```

```python
model = YOLO("C:/Users/it shop/Downloads/Telegram Desktop/yolov8n-seg.pt")  
results = model("E:/Senior Term1/New folder/SEG.jpg")  
```

```python
result_img = results[0].plot()
plt.imshow(result_img)
plt.axis('off')
plt.show()
```

```python
cap = cv2.VideoCapture("E:/Senior Term1/New folder/segm.mp4")

out = cv2.VideoWriter("C:/Users/it shop/runs/segment/output_segmented_video.mp4", cv2.VideoWriter_fourcc(*'mp4v'), cap.get(cv2.CAP_PROP_FPS), 
                      (int(cap.get(3)), int(cap.get(4))))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret: break
        
    result_img = model(frame)[0].plot()  
    out.write(result_img)  
    cv2.imshow("Processed Frame", result_img)  

cap.release()
out.release()

```
