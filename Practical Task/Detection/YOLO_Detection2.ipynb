```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO("C:/Users/it shop/Downloads/yolov8n.pt")
results = model(source="E:/Senior Term1/New folder/car2.mp4", stream=True, show=True, conf=0.4, save=True)

for result in results:
    frame = result.plot()  
    cv2.imshow("Processed Frame", frame) 
```
