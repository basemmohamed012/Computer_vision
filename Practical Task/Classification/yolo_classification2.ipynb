```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
model = YOLO("yolov8n-cls.pt")
results = model("E:/Senior Term1/New folder/imggg.jpg")

probs = results[0].probs
print("Class probabilities:", probs.data)
```

```python
result_img = results[0].plot()
plt.imshow(result_img)
plt.axis('off')
plt.show()
```

```python
import cv2
```

```python
cap = cv2.VideoCapture("E:/Senior Term1/New folder/segm.mp4")

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("output_video.mp4", fourcc, 30.0, (int(cap.get(3)), int(cap.get(4))))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    
    probs = results[0].probs
    print("Class probabilities:", probs.data)

    result_frame = results[0].plot()
    
    cv2.imshow("Processed Frame", result_frame)
    out.write(result_frame)

cap.release()
out.release()
```
