```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO("yolov8n.pt")
results = model(
    source=r"WhatsApp Video 2024-12-14 at 02.43.49_586028f3.mp4",
    stream=True,
    show=True,
    conf=0.4,
    save=True
)


for result in results:
    frame = result.plot()  
    cv2.imshow("Processed Frame", frame)  #    OpenCV
    
```
